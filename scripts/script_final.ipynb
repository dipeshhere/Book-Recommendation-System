Model Building- Popularity Based Approach




```python
ratings.head(2)
```

```python
ratings['user_id'].value_counts()
```

Collabrative filtering based approach:

1.Users with 200+ ratings == Knowledgeable users

2.Books with 50=> ratings == Popular/Famous Books


```python
x=ratings['user_id'].value_counts()>200
```

```python
x[x].shape
```

```python
y=x[x].index #we will the userid's index
```

```python
y
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings_with_books=ratings.merge(books,on='ISBN')
```

```python
ratings_with_books.shape
```

```python
number_rating=ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating':'number of ratings'},inplace=True)
```

```python
final_rating=ratings_with_books.merge(number_rating, on='title')
```

```python
final_rating.head(2) #has rating , book details, no. of ratings
```

```python
final_rating=final_rating[final_rating['number of ratings']>=50]
```

```python
final_rating.shape #61k are users wiht minimum 200 ratings and books who had minimum 50 ratings
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)  #Drop users with same user wiht same books
```

```python
final_rating.shape #Almost 2000 rows of duplicates valuse
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
#converting into sparce matrix
```

```python
from scipy.sparse import csr_matrix  #To only consider that values which is non-zero
book_sparse=csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
def recommend_book(book_name):
    # Find the book index safely
    matching_indices = np.where(book_pivot.index == book_name)[0]
    
    # Check if book exists
    if len(matching_indices) == 0:
        return None  # Book not found
    
    book_id = matching_indices[0]
    
    # Get recommendations
    distances, suggestions = model.kneighbors(
        book_pivot.iloc[book_id, :].values.reshape(1, -1), 
        n_neighbors=6
    )
    
    # Extract book names (skip the first one as it's the input book itself)
    books = book_pivot.index[suggestions[0]]
    
    return books[1:]  # Return recommendations excluding the input book
```
