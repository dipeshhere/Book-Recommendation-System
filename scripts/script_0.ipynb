```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name.head(2)
```

```python
num_rating_df=ratings_with_name.groupby("title").count()[['rating']].reset_index() #Gives no. of rating
num_rating_df.rename(columns={'rating':'num_rating'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('title')['rating'].mean().reset_index()
avg_rating_df.rename(columns={'rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
#Merging these 2 tables
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='title')
popular_df
```

Top 50 books based on average rating with a condition that there a around 250 rating for that book 

Collabrative filtering based approach:
1.Users with 200+ ratings == Knowledgeable users
2.Books with 50=> ratings == Popular/Famous Books

```python
x=ratings_with_name.groupby('user_id').count()['rating']>200
pdu=x[x].index #Boolean Indexing
```

```python
filtered_rating=ratings_with_name[ratings_with_name['user_id'].isin(pdu)]
```

Now, filtering on the basis of Books

```python
y=filtered_rating.groupby('title').count()['rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt=final_ratings.pivot_table(index='title',columns='user_id',values='rating')
pt
```

We have 706 Users and 810 Books

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores.shape #Similarity score of books with each other 
```

```python
similarity_scores
```

```python
# def recommend(book_name):
#     index=np.where(pt.index==book_name)[0][0] #index of the book 
#     similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]

#     for i in similar_items:
#         print(pt.index[i[0]])
    
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('title')['title'].values))
        item.extend(list(temp_df.drop_duplicates('title')['author'].values))
        # item.extend(list(temp_df.drop_duplicates('title')['url_medium'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('Harry Potter and the Prisoner of Azkaban (Book 3)')
```
